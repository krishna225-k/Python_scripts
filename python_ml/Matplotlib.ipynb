```python
import matplotlib.pyplot as plt
x_values = list(range(1000))
squares = [x**2 for x in x_values]
plt.scatter(x_values,squares,c=squares,cmap=plt.cm.Blues,edgecolors='none',s=10)
plt.scatter(x_values[0],squares[0],c='green',edgecolors='none',s=100)
plt.scatter(x_values[-1],squares[-1],c='red',edgecolors='none',s=100)

# Add features to graphs 
plt.title('Square Numbers',fontsize=24)
plt.xlabel('values',fontsize=24)
plt.ylabel('squares of values',fontsize=18)
# Hiding the x and y axies scaling 
plt.axes().get_xaxis().set_visible(True)
plt.axes().get_yaxis().set_visible(True)
# tick parameters 
plt.tick_params(axis='both',which='major',labelsize='large')
plt.axis([0,1100,0,1100000])
plt.figure(dpi=128,figsize=(10,6))
plt.show()
plt.savefig('squares.png')
```

Multiple plots

```python
import matplotlib.pyplot as plt

x_values = list(range(11))
squares = [x**2 for x in x_values]
cubes = [x**3 for x in x_values]
plt.scatter(x_values,squares, c='blue',edgecolors='none',s=10)
plt.scatter(x_values,cubes,c='red',edgecolors='none',s=20)

plt.axis([0,11,0,1100])
plt.show()
plt.fill_between(x_values, cubes,squares,facecolor='blue',alpha=0.5)
plt.savefig('multiple.png')
```

Density nad Coutour plots

Some times it is useful to display three-dimenstional data in two dimentionas using contour or color-coded 
There three matplotlib function can be helpful for task
1. plt.contour
2. plt.contourf
3. plt.imshow

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Visualizing Three dimensional function
======================================

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x, y)
Z = f(X,Y)
print(X)
```

```python
print(Y)
```

```python
'''Parameters
----------
X, Y : 
Z : array-like(N, M)
    The height values over which the contour is drawn.

levels : int or array-like, optional
    Determines the number and positions of the contour lines / regions.
Other Parameters
----------------
corner_mask : colors : alpha : cmap : norm : `vmin, vmax : origin : extent : locator : extend : xunits, yunits : antialiased : Nchunk : linewidths : linestyles : hatches : 
'''
plt.contour(X, Y, Z, colors = 'black') # Nagitive values represent by the dashed lines
                                        # Positive values represent by the solid lines
```

```python
plt.contour(X, Y, Z,20, cmap = 'RdGy')
plt.colorbar()
#cmap = Red and Gray
```

```python
'''The spaces between the lines may be a bit distracting, We can overcoming by filled contour plot:'plt.contour\''''
```

```python
plt.contourf(X,Y,Z, 20, cmap='RdGy')
plt.colorbar()# it's show the intensity of color on graph
```

```python
'''plt.imshow(
    ['X', 'cmap=None', 'norm=None', 'aspect=None', 'interpolation=None', 'alpha=None', 'vmin=None', 
    'vmax=None', 'origin=None', 'extent=None', 'shape=None', 'filternorm=1', 'filterrad=4.0', 'imlim=None', 'resample=None', 
    'url=None', '*', 'data=None', '**kwargs'],'''

plt.imshow(Z, extent=[0,5,0,5],origin='left',cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

```python
#Combining both image plots and contour plots
contours = plt.contour(X,Y,Z ,3, colors = 'black')
plt.clabel(contours,inline=True,fontsize =5)
plt.imshow(Z, extent=[0,5,0,5],origin='upper',cmap='RdGy',alpha=0.9)
plt.colorbar()
```
